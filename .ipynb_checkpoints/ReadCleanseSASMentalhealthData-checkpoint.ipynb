```python
import pandas as pd
import numpy as np
import os
```

```python
MentalHealthCSV = os.path.join('raw_data', 'LLCP2016.XPT')

mentalHealth_df = pd.read_sas(MentalHealthCSV)

#df = pd.read_sas('LLCP2016.XPT')



mentalHealth_df.head()
```

```python
mentalHealth_df.columns.values.tolist()
```

```python
# To pull needed columns only 


reduced_mental_health_data = mentalHealth_df.loc[:, ['IDATE','_STATE', 'SEX', '_AGEG5YR', '_PRACE1', 'EMPLOY1', 'MSCODE',
                                        'MENTHLTH',
                                        '_RFHLTH',
                                        '_HCVU651', 'SLEPTIM1','CVDINFR4', 
                                        'CVDCRHD4', 'CVDSTRK3', 'ASTHNOW', 'CHCSCNCR', 'CHCOCNCR',
                                        'CHCCOPD1',
                                        '_DRDXAR1','ADDEPEV2', 'CHCKIDNY', 'DIABETE3', 
                                        'PREGNANT', 'QLMENTL2', 'QLSTRES2', '_RFSMOK3', 
                                        '_RFDRHV5', '_TOTINDA', '_INCOMG', '_BMI5CAT', 
                                        'EMTSUPRT', 'LSATISFY']]

#New added:
#_RFHLTH',
  # '_HCVU651', '_DRDXAR1'
    
#    'QLMENTL2'
#'QLSTRES2', '_RFSMOK3', '_TOTINDA' (Phyicall Activity), '_INCOMG'

reduced_mental_health_data.head()
```

```python
reduced_mental_health_data["IDATE"] = reduced_mental_health_data["IDATE"].str.decode("utf-8")
reduced_mental_health_data["SEX"] = reduced_mental_health_data["SEX"].replace({1.0:"M",2.0:"F", None:"NA"})


reduced_mental_health_data["_AGEG5YR"]=reduced_mental_health_data["_AGEG5YR"].replace({1.0:"18-24",2.0:"25-29",3.0:"30-34",4.0:"35-39",5.0:"40-44",
                                                  6.0:"45-49",7.0:"50-54",8.0:"55-59",9.0:"60-64",10.0:"65-69",
                                                  11.0:"70-74",12.0:"75-79",13.0:"80+",14.0:"NA",None:"NA"})
reduced_mental_health_data["EMPLOY1"]=reduced_mental_health_data["EMPLOY1"].replace({1.0:"Employed", 2.0:"Self-Employed", 3.0:"Out of work",
                                               4.0:"Out of work", 5.0:"Homemaker", 6.0:"Student",
                                               7.0:"Retired", 8.0:"Unable to work", 9.0:"NA",None:"NA"})
```

```python
reduced_mental_health_data["_PRACE1"]=reduced_mental_health_data["_PRACE1"].replace({1.0:"White",2.0:"Black or African American",3.0:"Asian",
                                               4.0:"Native Hawaiian or Other Pacific Islander",
                                               5.0:"American Indian, Alaska Native",6.0:"Other",
                                               7.0:"NA", 8.0:"Multiracial",9.0:"NA"})
reduced_mental_health_data["MENTHLTH"]=reduced_mental_health_data["MENTHLTH"].replace({88:0,77:"NA",99:"NA",None:"NA"})
reduced_mental_health_data["_RFHLTH"]=reduced_mental_health_data["_RFHLTH"].replace({1:"Good",2:"Poor",9:"NA",None:"NA"})
#reduced_mental_health_data["HLTHPLN1"]=reduced_mental_health_data["HLTHPLN1"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
```

```python
#Chronic Health Disease 
reduced_mental_health_data["CVDINFR4"]=reduced_mental_health_data["CVDINFR4"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["CVDCRHD4"]=reduced_mental_health_data["CVDCRHD4"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["CVDSTRK3"]=reduced_mental_health_data["CVDSTRK3"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["ASTHNOW"]=reduced_mental_health_data["ASTHNOW"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["CHCSCNCR"]=reduced_mental_health_data["CHCSCNCR"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["CHCOCNCR"]=reduced_mental_health_data["CHCOCNCR"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["CHCCOPD1"]=reduced_mental_health_data["CHCCOPD1"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["_DRDXAR1"]=reduced_mental_health_data["_DRDXAR1"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
```

```python
reduced_mental_health_data["ADDEPEV2"]=reduced_mental_health_data["ADDEPEV2"].replace({1:"Y", 2:"N", 7:"NA", 9:"NA",None:"NA"})

reduced_mental_health_data["CHCKIDNY"]=reduced_mental_health_data["CHCKIDNY"].replace({1:"Y", 2:"N", 7:0, 9:0,None:0})
reduced_mental_health_data["DIABETE3"]=reduced_mental_health_data["DIABETE3"].replace({1:"Y", 2:"N", 3:"N", 4:"N", 7:0, 9:0,None:0})
```

```python
reduced_mental_health_data["PREGNANT"]=reduced_mental_health_data["PREGNANT"].replace({1:"Y", 2:"N", 7:"NA", 9:"NA",None:"NA"})
reduced_mental_health_data["_BMI5CAT"]=reduced_mental_health_data["_BMI5CAT"].replace({4:"Obese",3:"Overweight",2:"Normal Weight",1:"Underweight",None:"NA"})
reduced_mental_health_data["QLMENTL2"]=reduced_mental_health_data["QLMENTL2"].replace({88:0,77:"NA",99:"NA",None:"NA"})
reduced_mental_health_data["_TOTINDA"]=reduced_mental_health_data["_TOTINDA"].replace({2:"N",1:"Y",9:"NA",None:"NA"})
reduced_mental_health_data["QLSTRES2"]=reduced_mental_health_data["QLSTRES2"].replace({88:0,77:"NA",99:"NA",None:"NA"})
```

```python
reduced_mental_health_data["_RFSMOK3"]=reduced_mental_health_data["_RFSMOK3"].replace({1:"Non-Smoker",2:"Smoker",9.0:"NA",None:"NA"})
                                                   #7.0:0,9.0:0,None:0})

reduced_mental_health_data["_RFDRHV5"]=reduced_mental_health_data["_RFDRHV5"].replace({1:"Non-Alcoholic",2:"Alcoholic",9.0:"NA",None:"NA"})

reduced_mental_health_data["MSCODE"]=reduced_mental_health_data["MSCODE"].replace({5:"Rural",3:"Urban",2:"Urban",1:"Urban",None:"Rural"})



```

```python
reduced_mental_health_data["LSATISFY"]=reduced_mental_health_data["LSATISFY"].replace({1: "Satisfied", 2: "Satisfied", 3: "Not-Satisfied", 4: "Not-Satisfied", 7: "NA", 9:"NA",None:"NA"})
reduced_mental_health_data["EMTSUPRT"]=reduced_mental_health_data["EMTSUPRT"].replace({7:"NA", 9:"NA", None:"NA"})

reduced_mental_health_data["_INCOMG"]=reduced_mental_health_data["_INCOMG"].replace({1:"<15000", 2:"15000-<25000",
                                                 3:"25000-<35000", 4:"35000-<50000",
                                                 5:"50000+",9:"NA",None:"NA"})

```

```python
reduced_mental_health_data.head()
```

```python
# Converting State numbers into State Codes

reduced_mental_health_data = reduced_mental_health_data.rename(columns={"_STATE":"fips"})


outputfilepath=os.path.join("raw_data","fipstostate.csv")
FIP_State=pd.read_csv(outputfilepath)
FIP_State.head()

```

```python
reduced_mental_health_data.head()
```

```python
reduced_mental_health_data = reduced_mental_health_data.merge(FIP_State,on="fips")
reduced_mental_health_data.drop("fips",axis=1,inplace=True)
reduced_mental_health_data["Chronic_Disease"]=""

reduced_mental_health_data.head()
```

```python
# Function to Aggregating Chronic Disease 

for index, row in reduced_mental_health_data.iterrows():
    chron_dep=0
    if row["CVDINFR4"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["CVDCRHD4"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["CVDSTRK3"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["ASTHNOW"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["CHCSCNCR"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["CHCOCNCR"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["CHCCOPD1"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["_DRDXAR1"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["CHCKIDNY"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    if row["DIABETE3"]=="Y":
        chron_dep+=1
    else:
        chron_dep+=0
    reduced_mental_health_data.set_value(index, "Chronic_Disease",chron_dep)


reduced_mental_health_data.drop(["CVDINFR4","CVDCRHD4","CVDSTRK3","ASTHNOW","CHCSCNCR",
          "CHCOCNCR","CHCCOPD1","_DRDXAR1","CHCKIDNY","DIABETE3"],axis=1,inplace=True)

reduced_mental_health_data.head()
```

```python
reduced_mental_health_data = reduced_mental_health_data.rename(columns={"_AGEG5YR":"Age_Range",
                                  "state_abbr":"State","SEX":"Gender",
                                  "MSCODE":"MetroStatusCode",
                                    "EMPLOY1":"Employment_Status",
                                  "MENTHLTH":"Mental_Health_Factor", 
                                  "_RFHLTH":"General_Health",
                                    "_HCVU651":"Health_Coverage",
                                  "ADDEPEV2":"Chronic_Depression",
                                  "QLACTLM2":"DISABLED",
                                  "SLEPTIM1":"Sleep_Hours",
                                    "QLMENTL2":"Depressed_Days",
                                  "QLSTRES2":"Stress_Days", 
                                  "_RFSMOK3":"Smoke_Frequency",
                                    "_RFDRHV5":"Alcohol_Frequency",
                                  "PREGNANT":"Pregnant",
                                    "IDATE":"Date","_PRACE1":"Race",
                                  "_TOTINDA":"Physical_Activity",
                                  "_INCOMG":"Income","_BMI5CAT":"BMI",
                                 "EMTSUPRT":"EmotionalSupport_Availability",
                                  "LSATISFY":"LifeSatisfaction_Level",
                                  

})

reduced_mental_health_data.head()
```

```python
outputfilepath=os.path.join("raw_data","BRFSS_2016.csv")

reduced_mental_health_data.to_csv(outputfilepath,index=False)
```
