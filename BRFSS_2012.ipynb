```python
import pandas as pd
import numpy as np
```

```python
DF_2012=pd.read_sas("LLCP2012.XPT")
```

```python
FIP_State=pd.read_csv("fipstostate.csv")
#FIP_State.set_index("fips",inplace=True)
FIP_State.head()
```

```python
DF_2012.head()
```

```python
NDF_2012=DF_2012[["IDATE","_STATE", "SEX", "_AGEG5YR", "ORACE2","CTYCODE1", "EMPLOY","MENTHLTH",
                    "POORHLTH","HLTHPLN1","CVDINFR4","CVDCRHD4","CVDSTRK3", "ASTHNOW","CHCSCNCR", 
                    "CHCOCNCR","CHCCOPD1","HAVARTH3","ADDEPEV2", "CHCKIDNY", "DIABETE3","PREGNANT", 
                    "QLACTLM2","QLMENTL2","QLSTRES2","SMOKDAY2", "AVEDRNK2", "MISNERVS","MISHOPLS",
                    "MISRSTLS","MISDEPRD","MISWTLES","MISNOWRK","MISTMNT","QLREST2", "EMTSUPRT", "LSATISFY"]]
```

```python
NDF_2012=NDF_2012.rename(columns={"_STATE":"fips"})
NDF_2012=NDF_2012.merge(FIP_State,on="fips")
```

```python
# Convert date from byte to string
NDF_2012["IDATE"]=NDF_2012["IDATE"].str.decode("utf-8")

# Convert Sex Field in DF from 1.0 to Male and 2.0 to Female
NDF_2012["SEX"]=NDF_2012["SEX"].replace({1.0:"M",2.0:"F"})


NDF_2012["_AGEG5YR"]=NDF_2012["_AGEG5YR"].replace({1.0:"18-24",2.0:"25-29",3.0:"30-34",4.0:"35-39",5.0:"40-44",
                                                  6.0:"45-49",7.0:"50-54",8.0:"55-59",9.0:"60-64",10.0:"65-69",
                                                  11.0:"70-74",12.0:"75-79",13.0:"80+",14.0:""})
NDF_2012["ORACE2"]=NDF_2012["ORACE2"].replace({1.0:"White",2.0:"Black or African American",3.0:"Asian",
                                               4.0:"Native Hawaiian or Other Pacific Islander",
                                               5.0:"American Indian, Alaska Native",6.0:"Other",
                                               7.0:"nan", 8.0:"Multiracial",9.0:"nan"})
NDF_2012["EMPLOY"]=NDF_2012["EMPLOY"].replace({1.0:"Employed", 2.0:"Self-Employed", 3.0:"Out of work",
                                               4.0:"Out of work", 5.0:"Homemaker", 6.0:"Student",
                                               7.0:"Retired", 8.0:"Unable to work", 9.0:"nan"})
NDF_2012["MENTHLTH"]=NDF_2012["MENTHLTH"].replace({88:0,77:"nan",99:"nan"})
NDF_2012["POORHLTH"]=NDF_2012["POORHLTH"].replace({88:0,77:"nan",99:"nan"})
NDF_2012["HLTHPLN1"]=NDF_2012["HLTHPLN1"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["CVDINFR4"]=NDF_2012["CVDINFR4"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["CVDCRHD4"]=NDF_2012["CVDCRHD4"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["CVDSTRK3"]=NDF_2012["CVDSTRK3"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["ASTHNOW"]=NDF_2012["ASTHNOW"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["CHCSCNCR"]=NDF_2012["CHCSCNCR"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["CHCOCNCR"]=NDF_2012["CHCOCNCR"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["CHCCOPD1"]=NDF_2012["CHCCOPD1"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["HAVARTH3"]=NDF_2012["HAVARTH3"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["ADDEPEV2"]=NDF_2012["ADDEPEV2"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["CHCKIDNY"]=NDF_2012["CHCKIDNY"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["DIABETE3"]=NDF_2012["DIABETE3"].replace({1:"Y", 2:"N", 3:"N", 4:"N", 7:"nan", 9:"nan"})
NDF_2012["PREGNANT"]=NDF_2012["PREGNANT"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["QLACTLM2"]=NDF_2012["QLACTLM2"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["QLMENTL2"]=NDF_2012["QLMENTL2"].replace({88:0,77:"nan",99:"nan"})
NDF_2012["QLSTRES2"]=NDF_2012["QLSTRES2"].replace({88:0,77:"nan",99:"nan"})
NDF_2012["SMOKDAY2"]=NDF_2012["SMOKDAY2"].replace({1.0:"Everyday",2.0:"Some days",3.0:"None",
                                                   7.0:"nan",9.0:"nan"})
NDF_2012["AVEDRNK2"]=NDF_2012["AVEDRNK2"].replace({77:"nan",99:"nan"})
NDF_2012["MISNERVS"]=NDF_2012["MISNERVS"].replace({1.0:"All", 2.0:"Most", 3.0:"Some", 4.0:"Little", 5.0:"None",
                                                  7.0:"nan", 9.0:"nan"})
NDF_2012["MISHOPLS"]=NDF_2012["MISHOPLS"].replace({1.0:"All", 2.0:"Most", 3.0:"Some", 4.0:"Little", 5.0:"None",
                                                  7.0:"nan", 9.0:"nan"})
NDF_2012["MISRSTLS"]=NDF_2012["MISRSTLS"].replace({1.0:"All", 2.0:"Most", 3.0:"Some", 4.0:"Little", 5.0:"None",
                                                  7.0:"nan", 9.0:"nan"})
NDF_2012["MISDEPRD"]=NDF_2012["MISDEPRD"].replace({1.0:"All", 2.0:"Most", 3.0:"Some", 4.0:"Little", 5.0:"None",
                                                  7.0:"nan", 9.0:"nan"})
NDF_2012["MISWTLES"]=NDF_2012["MISWTLES"].replace({1.0:"All", 2.0:"Most", 3.0:"Some", 4.0:"Little", 5.0:"None",
                                                  7.0:"nan", 9.0:"nan"})
NDF_2012["MISNOWRK"]=NDF_2012["MISNOWRK"].replace({88:0,77:"nan",99:"nan"})
NDF_2012["MISTMNT"]=NDF_2012["MISTMNT"].replace({1:"Y", 2:"N", 7:"nan", 9:"nan"})
NDF_2012["QLREST2"]=NDF_2012["QLREST2"].replace({88:0,77:"nan",99:"nan"})


# Renaming the columns
NDF_2012=NDF_2012.rename(columns={"_AGEG5YR":"AGE","ORACE2":"RACE","CTYCODE1":"COUNTY CODE",
                                    "EMPLOY":"EMPLOYMENT STATUS","MENTHLTH":"MENTAL HLTH", "POORHLTH": "POOR HLTH",
                                    "HLTHPLN1":"HEALTH PLAN","ADDEPEV2":"HAD DEPPRESSION","QLACTLM2":"DISABLED",
                                    "QLMENTL2":"DEPRESSED DAYS","QLSTRES2":"STRESSED DAYS", "SMOKDAY2":"CIGARETTE CONSUMPTION",
                                    "AVEDRNK2":"AVG DRINKS(MONTH)","QLREST2":"SLEEPLESSDAYS","state_abbr":"STATE",
                                    "IDATE":"DATE"})
NDF_2012.head()
```
